```python
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
c6 = []
for ind,i in enumerate(l):
    print(ind)
    a = cv2.imread(os.path.join('../Data/CBIS-DDSM_classification_orient/',i))
    a = cv2.resize(a,(256,320))
    c = np.zeros((288,256,3))
    c[:,:,0] = a[16:16+288,:,0]
    c[:,:,1] = a[16:16+288,:,1]
    c[:,:,2] = a[16:16+288,:,2]
    c1.append(c[:,:,0].mean())
    c2.append(c[:,:,1].mean())
    c3.append(c[:,:,2].mean())
    c4.append(c[:,:,0].std())
    c5.append(c[:,:,1].std())
    c6.append(c[:,:,2].std())
    
    cv2.imwrite('../Data/CBIS-DDSM_classification_orient_cropped/'+i,a)
    
```

```python
c1 = np.array(c1)
c2 = np.array(c2)
c3 = np.array(c3)
c4 = np.array(c4)
c5 = np.array(c5)
c6 = np.array(c6)
```

```python
min_l = []
max_l = []

for ind,i in enumerate(l):
    print(ind)
    a = cv2.imread(os.path.join('../Data/mask_orient_cropped/',i),0)
    c = np.zeros((288,256))
    b = a
    b[b<0.2] = 0
    b[b>0.2] = 1
    r,c = np.where(b==1)
    r_m = int(r.mean())
    c_m = int(c.mean())
    distance_mask = np.zeros((288,256))
    for i1 in range(288):
        for i2 in range(256):
            distance_mask[i1,i2] = np.sqrt((i1-r_m)**2+(i2-c_m)**2)
    distance_mask = distance_mask/distance_mask.max()
    cv2.imwrite('../Data/distance_mask_orient_cropped/'+i,distance_mask*255)
    
```
