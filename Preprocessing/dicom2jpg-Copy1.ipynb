```python
import pydicom                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                       
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
import cv2
import skimage
import skimage.io as io
from natsort import natsorted
import pdb
```

```python
read_path = '../../Data/CBIS-DDSM/'
write_path = '../../Data/CBIS-DDSM_ex_1/'
```

```python
os.mkdir(write_path)
```

```python
file_list = natsorted(os.listdir(read_path))

for ii in range(len(file_list)):
    #pdb.set_trace()
    print(ii)
    image_name = file_list[ii]
    if image_name[-1] == '1' or image_name[-1] == '2' or image_name[-1] == '3' or image_name[-1] == '4' or \
    image_name[-1] == '5' or image_name[-1] == '6' or image_name[-1] == '7' or image_name[-1] == '8':
        continue
    else:
        f1 = os.path.join(read_path,image_name)
        l1 = os.listdir(f1)
        
        f2 = os.path.join(f1,l1[0])
        l2 = os.listdir(f2)
        
        f3 = os.path.join(f2,l2[0])
        l3 = os.listdir(f3)
        
        f4 = os.path.join(f3,l3[0])
    
        mammo_file = pydicom.read_file(f4).pixel_array
        io.imsave(os.path.join(write_path,image_name)+'.jpg',mammo_file/mammo_file.max())


        for i in range(7):
            ind = i+1
            
            m1 = f1+'_'+str(ind)
            if os.path.isdir(m1):
                for a in sorted(os.listdir(m1)):
                    m2 = os.path.join(m1,a)
                    for b in sorted(os.listdir(m2)):
                        m3 = os.path.join(m2,b)
                        for c in sorted(os.listdir(m3)):
                            if c[-3:] == "dcm":
                                mammo_file_1 = pydicom.read_file(os.path.join(m3,c)).pixel_array
                                if np.prod(mammo_file_1.shape) < np.prod(mammo_file.shape) - 500:
                                    io.imsave(os.path.join(write_path,image_name+'_'+str(ind))+'@patch.jpg',mammo_file_1/mammo_file_1.max())
                                else:
                                    io.imsave(os.path.join(write_path,image_name+'_'+str(ind))+'@mask.jpg',mammo_file_1/mammo_file_1.max())
                
    
```
